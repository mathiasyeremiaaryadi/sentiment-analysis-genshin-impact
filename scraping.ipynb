# Import Library

```python
from google_play_scraper import reviews, reviews_all, Sort
import pandas as pd
```

# Google Play Scrapping

```python
all_reviews = reviews_all(
    app_id='com.miHoYo.GenshinImpact',
    lang='en',
    sort=Sort.MOST_RELEVANT,
    count=10000,
)
```

```python
print('Example review data: ', all_reviews[1])
print('Size of reviews: ', len(all_reviews))
```

# Store Dataset

Menyimpan data hasil scrapping ke dalam file csv

```python
df = pd.DataFrame(all_reviews)
df.head()
```

```python
df.shape
```

```python
df.to_csv('dataset_raw.csv', index=False)
```
